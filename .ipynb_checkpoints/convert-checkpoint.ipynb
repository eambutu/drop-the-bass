```python
from struct import pack, unpack
import pyaudio
import wave
import sys
import time
import numpy as np
from scipy.fftpack import dct, idct
import matplotlib.pyplot as plt
from scipy.signal import butter, lfilter
```

```python
filename = "nevergonna"

wf = wave.open('./' + filename + ".wav", 'rb')
orig = []
recon = []
CHUNK = 8192
data = wf.readframes(CHUNK)
tot_time = 0
num_chunks = 0
chunk_freqs = []

def unsigned_to_signed(x):
    return x if x < (1 << 16-1) else x - (1 << 16)

def signed_to_unsigned(x):
    return x & 0xffff

def convert(data, flip=True):
    global tot_time, num_chunks, orig, recon, last_chunk_freqs
    start = time.time()
    data = bytearray(data)
 
    data_16 = [data[i] + data[i+1] * 256 for i in range (0,len(data), 2)]
    signed_data = np.clip([unsigned_to_signed(x) for x in data_16], -32767, 32767)

    freqs = np.fft.fft(signed_data)
    
    freqs = shift(freqs)
    
    chunk_freqs.append(freqs)
        
    reconstruct = np.clip([int(x) for x in np.real(np.fft.ifft(freqs))], -32767, 32767)
    unsigned_reconstruct = [signed_to_unsigned(x) for x in reconstruct]
    
    orig.append(data_16)
    recon.append(unsigned_reconstruct)

    output = b''
    for i in range(0, len(data_16)):
        output += pack('H', data_16[i])
        output += pack('H', unsigned_reconstruct[i])

                
    end = time.time()
    tot_time += end - start
    num_chunks += 1
    return output

wvData = b''
while data != b'':
    temp = convert(data, flip=True)
    wvData += temp
    data = wf.readframes(CHUNK)

wf_out = wave.open('./' + filename + "_trip.wav", 'w')
wf_out.setparams((2, 2, wf.getframerate(), 0, 'NONE', 'not compressed'))
wf_out.writeframes(wvData)
wf_out.close()
```

```python
%matplotlib nbagg
index = 20

fig = plt.figure()
ax1 = fig.add_subplot(211)
ax1.plot([unsigned_to_signed(x) for x in orig[index]], "b-s")
ax1.plot([unsigned_to_signed(x) for x in recon[index]], "r-s")
ax2 = fig.add_subplot(212)
ax2.plot(chunk_freqs[index], "b-s")


def shift(freqs):
    mid_high = len(freqs)/2 + 1
    return np.append(np.append(np.append(np.append([freqs[0]], [freqs[0]]), [freqs[1:mid_high-1]]),  
                               [freqs[mid_high+1:]]), [[freqs[-1]]])

ax2.plot(shift(chunk_freqs[index]), "r-s")
```
